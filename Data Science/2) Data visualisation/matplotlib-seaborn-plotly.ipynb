## Importation de la bibliothèque  <a class="anchor" id="section_2_1"></a>

On importe la bibliothèque ```matplotlib.pyplot``` souvent renommée ```plt``` dont la documentation est disponible [ici](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html)

````python
import matplotlib.pyplot as plt
````

## Tracé basique  <a class="anchor" id="section_2_2"></a>

La fonction ```plot``` permet de tracer un tracé basique. Elle prend en argument une séquences de données pour les abscices (```X```) et une séquence de données pour les ordonnées (```Y```).

````python
import numpy as np
import pandas as pd
from math import exp

# création d'une séquence de d'abscices
X = np.linspace(1, 10, num=100)

# création d'une séquence en ordonnées
Y = [exp(elt) for elt in X]
````

````python
# tracer X en fonction de Y
plt.plot(X,Y)
# affichage du graphe
plt.show()
````

* <span class="text-warning"> $\textbf{Remarque}$: Si on ne travaille pas sur un Jupyter notebook, il faut ajouter la commande ```plt.show()```pour afficher la figure </span> 

### Edition de paramètres  <a class="anchor" id="section_2_2_1"></a>

* <span class="text-muted"> Ajouter le titre "Fonction exponentielle" </span> 
* <span class="text-muted"> Ajouter le titre "Axe X" pour l'axe des abscices  </span> 
* <span class="text-muted"> Ajouter le titre "Axe Y" pour l'axe des ordonnées  </span> 
* <span class="text-muted"> Délimiter l'affichage graphique pour $X \in [8,10]$ et $Y \in [5000,15 000]$   </span> 

````python
# tracer de la fonction
plt.plot(X,Y) 
# ajouter un titre
plt.title('Fonction exponentielle')
# ajouter un label à l'axe des abscices
plt.xlabel('Axe X')
# ajouter un label à l'axe des ordonnées
plt.ylabel('Axe Y')
# délimiter l'affichage sur l'abscices
plt.xlim(8,10)
# délimiter l'affichage sur les ordonnées
plt.ylim(5000, 15000)
# affichage du graphe
plt.show()
````

* <span class="text-muted"> Tracer la courbe en pointillés, orange (```#FF8C00```) </span>


````python
# tracer de la courbe
plt.plot(X, Y, color="#FF8C00", ls=':') 
# ajout du titre
plt.title('Fonction exponentielle')
# ajouter un label à l'axe des abscices
plt.xlabel('Axe X')
# ajouter un label à l'axe des ordonnées
plt.ylabel('Axe Y')
# affichage du graphe
plt.show()
````

* <span class="text-muted"> Créer plusieurs fenêtres grâce à subplot() </span>

````python
# délimitation des subplots
fig, axes = plt.subplots(nrows=2, ncols=2)
# tracer des 4 figures
axes[0][0].plot(X, Y)
axes[0][1].plot(X, Y, color="red", ls=':')
axes[1][0].plot(X, Y, color="green", ls='--')
axes[1][1].plot(X, Y, color="purple", ls='-.')
````

## Tracé d'un scatter plot  <a class="anchor" id="section_3_2"></a>

La fonction ```scatterplot``` permet de tracer des nuages de points

### Le paramètre ```hue```  <a class="anchor" id="section_3_2_1"></a>

Il est également possible de "rajouter une autre variable catégorielle" : on colorie les données par rapport aux valeurs de cette nouvelle variable.

# Plotly

Plotly dont la documentation se trouve [ici](https://plotly.com/python), permet de réaliser des visualisations graphiques interactives.

* <span class="text-muted"> Importer la bibliothèque ```plotly``` sous le nom ```px``` </span>

````python
import plotly.express as px
````

* <span class="text-muted"> Importation du data set ```gapminder```</span>

````python
df5 = px.data.gapminder()
df5.tail()
````

* Tracé de la population par continent au fil des années

````python
px.bar(df5, x="continent", y="pop", color="continent",
      animation_frame="year", animation_group="country", range_y=[0,4000000000])
````

* Tracé de l'espérance de vie en fonction du PIB par pays au fil des années

````python
px.scatter(df5, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
````
