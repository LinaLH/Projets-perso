# Introduction au Machine Learning

## Catégories de Machine Learning

* En ```Machine Learning``` il y'a principalement deux catégories de méthodes d'apprentissage :

sup_unsup.png

### Supervised Learning

* L’apprentissage supervisé est une approche de création d’intelligence artificielle (IA), où un algorithme informatique est formé sur les données d’entrée qui ont été ```étiquetées``` pour un produit particulier.

* Le modèle est entraîné jusqu’à ce qu’il puisse détecter les relations entre les données d’entrée et les étiquettes de sortie, ce qui lui permet de produire des résultats d’étiquetage précis lorsqu’on lui présente des données jamais vues auparavant.

L'apprentissage supervisé se divise en 2 parties :

intosl.png

* Avec la régression, nous utilisons des variables ```explicatives``` afin de prédire des variables ```continues```. Par exemple, avec différents paramètres sur une maison (taille, localisation, âge...) nous pouvons prédire son prix.

* Avec la classification, nous créons des ```catégories``` afin de prédire une classe. Par exemple, avec quelques informations sur un article (titre, mots-clés...) nous pouvons prédire la catégorie de cet article (économie, politique, sport...).

### Unsupervised Learning

* Contrairement à l'apprentissage supervisé, l'apprentissage non supervisé nécessite des données ```non étiquettées```. L’algorithme se débrouille pour déterminer la classification.

* Il s'agit donc de découvrir les structures sous-jacentes à ces données non étiquetées. Puisque les données ne sont pas étiquetées, il est impossible à l'algorithme de calculer de façon certaine un score de réussite.

## Etapes du Machine Learning

Afin de pouvoir construire un modèle de Machine Learning et obtenir des résultats satisfaisants il faut passer par un bon nombre d'étapes :

### Récupération des données

Plusieurs manières différentes existent pour récupérer des données :

* Soit on trouve des données déjà existantes sur ```Kaggle``` par exemple comme on l'as vu précedemment ;

### Preprocessing des données

* Une fois les données récupérées sous forme de ```dataset``` en csv ou autre il faut les nettoyer et effectuer un ensemble d'opérations qui varient selon le modèle que l'on va utiliser ou les résultats que l'on souhaite obtenir.

* On verra dans les futures formations les différents preprocessing sur du texte pour des modèles de ```NLP``` ou sur des images pour des modèles de ```Computer Vision``` par exemple.

* En règle général on a besoin de tester notre modèle, on va donc diviser nos données en ```train``` et ```test``` data, la plus grosse partie va servir à entrâiner notre modèle tandis que la plus petite partie va servir à tester notre modèle dessus.

### Construction du modèle

* Maintenant que nos données sont prêtes à être présenter à notre modèle il va falloir... le construire !

* Pour ce faire en règle générale on va utiliser des libraries Python déjà existantes et assez connu pour le Machine Learning :

  * ```Scikit-Learn``` pour de la Classification ;
  * ```NLTK```, ```Gensim``` pour du Natural Langage Processing ;
  * ```OpenCV```, ```OpenPose``` pour de la Computer Vision ;
  * ```Keras```, ```TensorFlow``` pour les Neural Networks.

### Entraînement du modèle

* C'est la partie qui prend le plus de temps d'éxecution et surtout qui consomme le plus de ressources. Quand le modèle est les données sont prêtes il faut laisser le modèle s'entraîner sur les ```train data``` pour qu'il puisse ensuite nous faire des prédictions sur les ```test data```.

* Il faut cependant faire attention à éviter les deux problèmes les plus courant en Machine Learning : ```Underfitting``` et ```Overfitting```.

under_over.png

* ```Overfitting``` signifie que nous avons «trop bien» entraîné le modèle et qu’il colle maintenant beaucoup trop étroitement aux données d'entraînement et fera donc des prédictions erronées sur des données sur lesquelles il n'a pas été entraîné.

* ```Underfitting``` signifie que nous avons «mal» entraîné le modèle et qu'il ne colle pas aux données d'entraînement et par conséquent il ne reflète pas les tendances des données.

* Des méthodes existent pour contourner ce genre de problème comme la ```Cross Validation``` qui réserve simplement une partie des données d'entraînement et les utilise pour tester le modèle tandis que les données restantes autres que celles réservées sont utilisées pour former le modèle. Parmi les types de ```Cross Validation``` le plus intuitif est sûrement le ```K-Fold``` :

kfold.png

### Prédictions avec le modèle

* Lorsqu'on se retrouve avec un modèle entraîné il faut le tester sur les données de validation pour voir s'il donne de bonnes prédictions.

* Pour certains cas d'usage assez généraux, il existe même des ```pre-trained model``` qu'on peut réutiliser comme base pour ensuite les adapter à toutes sortes d'applications très spécifiques c'est ce qu'on appelle le ```Transfert Learning```.

* Le modèle OpenSource le plus celèbre est ```Yolo``` : un modèle de reconnaissance d'objet et de personnes

### Evaluation du modèle

* Afin de décider si notre modèle est satisfaisant il faut établir des métriques pour l'évaluer et aussi expliquer pourquoi il fait telle ou telle prédiction c'est ce qu'on appelle le ```Machine Learning Explainability```.

* Il existe un très grand nombre de métriques assez diverses et variées il faut donc adapter en fonction du modèle que l'on souhaite évaluer.

* Par exemple pour un modèle de classification parmi les métriques les plus significatives on peut citer :

  * Precision : taux de prédictions positives correctes

  $$\frac{True Postive}{True Postive + False Postive}$$

  * Recall : taux de positifs correctement prédits

  $$\frac{True Postive}{True Postive + False Negative}$$

  * F1-Score : capacité d'un modèle à bien prédire les individus positifs
  
  $$\frac{2 * precison * recall}{precision + reacall}$$

  * Confusion Matrix : chaque ligne correspond à une classe réelle, chaque colonne correspond à une classe prédite

## Linear Model Theory

### Simple Linear Model

L’équation d'une régression linéaire simple est donnée par :

$$y=\alpha+\beta x+\varepsilon$$


* $y$ est l'objectif

* $x$ est la caractéristique

* $\alpha$ et $\beta$ sont des paramètres de l'équation linéaire. **Notre objectif est de les trouver**

* $\varepsilon$ est l'erreur

On peut réécrire l'équation :

$$\tilde{y}=\tilde{\alpha}+\tilde{\beta} x$$

Avec :

$$\varepsilon=y-\tilde{y}$$

$\varepsilon$ représente la distance entre les vraies valeurs et les valeurs théoriques, pour notre problème on suppose que $\varepsilon \sim \mathcal{N}\left(0, 1\right)$ 


````python
from sklearn.datasets import make_regression
XX, yy = make_regression(n_features = 1, noise=50, random_state=543)

def linear_regression(X, y, m_current=0, b_current=0, epochs=1000, learning_rate=0.0001):
    N = float(len(y))
    
    ms = [m_current] # a place to store our m values
    bs = [b_current] # a place to store our b values
    
    for i in range(epochs):
        y_current = (m_current * X) + b_current
        cost = sum([data**2 for data in (y-y_current)]) / N
        m_gradient = -(2/N) * sum(X * (y - y_current))
        b_gradient = -(2/N) * sum(y - y_current)
        m_current = m_current - (learning_rate * m_gradient)
        b_current = b_current - (learning_rate * b_gradient)
        
        ms.append(m_current)
        bs.append(b_current)
        
    return ms, bs

# run the linear regression and store the m and b values
ms, bs = linear_regression(XX.ravel(), yy, 
                           epochs=30, learning_rate=0.08)
# build the lines, mx + b for all ms X's and b's
y_lines = (ms * XX + bs).T # transpose for easier indexing

import plotly.graph_objects as go
# create the scatter plot
points = go.Scatter(x=XX.flatten(), y=yy, mode='markers')
# create initial line
line = go.Scatter(x=XX.flatten(), y=y_lines[0])
# create a layout with out title (optional)
layout = go.Layout(title_text="Gradient Descent Animation")
# combine the graph_objects into a figure
fig = go.Figure(data=[points, line])
                                 
# to see what we have so far
#fig.show()

# create a list of frames
frames = []
# create a frame for every line y
for i in range(len(y_lines)):
    # update the line
    line = go.Scatter(x=XX.flatten(), y=y_lines[i])
    # create the button
    button = {
        "type": "buttons",
        "buttons": [
            {
                "label": "Play",
                "method": "animate",
                "args": [None, {"frame": {"duration": 20}}],
            }
        ],
    }
    # add the button to the layout and update the 
    # title to show the gradient descent step
    layout = go.Layout(updatemenus=[button], 
                       title_text=f"Principe (iteration {i})")
    # create a frame object
    frame = go.Frame(
        data=[points, line], 
        layout=go.Layout(title_text=f"Principe: iteration {i}")
    )
# add the frame object to the frames list
    frames.append(frame)
    
# combine the graph_objects into a figure
fig = go.Figure(data=[points, line],
                frames=frames,
                layout = layout)
                                 
# show our animation!
fig.show()
````

### Calculate efficiency

#### Coefficient of determination

* Le coefficient de détermination $R^{2}$ (ou $r^{2}$) détermine la capacité d’un modèle à prédire ou à expliquer un résultat dans le paramètre de régression linéaire.

* Plus précisément, $R^{2}$ indique la proportion de la variance de la variable dépendante $Y$ qui est prédit ou expliquée par la régression linéaire et la variable prédictive $X$, aussi connue sous le nom de variable indépendante.

$$
R^{2}=1-\frac{\sum_{i=1}^{n}\left(y_{i}-\hat{y}_{i}\right)^{2}}{\sum_{i=1}^{n}\left(y_{i}-\bar{y}\right)^{2}}
$$

#### MSE and RMSE

**MSE** : L’erreur quadratique moyenne représente la moyenne de la différence quadratique entre les valeurs initiales et prédit de l’ensemble de données. Il mesure la variance des résidus.

$$
M S E=\sqrt{\sum_{i=1}^{n} \frac{\left(\hat{y}_{i}-y_{i}\right)^{2}}{n}}
$$

**RSME** : La racine carré de l'erreur quadratique moyenne est mesure l’écart-type des résidus à l'instrat de l'écart.

$$ RMSE=\sqrt{M S E}=\sqrt{\frac{1}{N} \sum_{i=1}^{N}\left(y_{i}-\hat{y}\right)^{2}}$$

### Importation

````python
import pandas as pd
import numpy as np
df = pd.read_csv('health_insurance.csv')
df.head(7)
````

Dataset information:

* **age** - the age of the customer (years)
* **sex** - the biological sex of the customer (male or female)
* **bmi** - the body mass index of the customer
* **children** - the number of children that the customer has
* **smoker** - whether the customer is a smoker or not
* **region** - the region that the customer lives in
* **logcharges** - the logarithm of the amount of money charged to the insurance company for the medical care of the customer (US dollars)

### Preprocessing

* <span class="text-muted"> Transform ```logcharges``` column into ```charges```   </span>

````python
df['charges'] = df['logcharges'].apply(lambda x : np.exp(x))
````

* <span class="text-muted"> Create a new column ```weight_status``` thanks to the following information   </span>

\begin{array}{|l|l|}
\hline \text { BMI } & \text { Weight Status } \\
\hline \text { Below } 18.5 & \text { Underweight } \\
\hline 18.5-24.9 & \text { Healthy Weight } \\
\hline 25.0-29.9 & \text { Overweight } \\
\hline 30.0 \text { and Above } & \text { Obesity } \\
\hline
\end{array}

````python
def status(x) :
  if (x < 18.5) : return ('Underweight')
  if (x > 18.5 and x < 24.9) : return ('Healthy Weight')
  if (x > 25 and x < 29.9) : return ('Overweight')
  else : return ('Obesity')
df['weight_status'] = df['bmi'].apply(status)
````

* <span class="text-muted"> Create two lists:   </span>
    * ```categorial_var```which contains: ```sex```, ```children```, ```smoker``` and```region```
    * ```continuous_var```which contains ```bmi```and ```charges```

````python
categorial_var = ['sex', 'children', 'smoker', 'region', 'weight_status']
df[categorial_var]
````

````python
continuous_var = ['bmi', 'charges']
df[continuous_var]
````

````python
df.drop(columns='logcharges', inplace=True)
````

### Encoding

* Regression model cannot analyze categorial variables, in order to solve that we have to encode categorial variables.

````python
from sklearn.preprocessing import LabelEncoder
for feature in categorial_var:
    df[feature] = LabelEncoder().fit_transform(df[feature])
df.head()
````

### Normalization

* Consider this data set containing two features: ```age``` and ```charges```. Where ```age``` ranges from 0 to 100, while income ranges from 0 to 64.

* These two features are in very different ranges. When we do further analysis, like multivariate linear regression, for example,```charges``` will intrinsically influence the result more due to its larger value.

* But this doesn’t necessarily mean it is more important as a predictor. So we normalize the data to bring all the variables to the same range

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[continuous_var] = scaler.fit_transform(df[continuous_var])
df.head()
````

### Features and Target

````python
y = df['charges']
X = df.drop(columns='charges')
````

### Linear Regression

* <span class="text-muted"> Import ```linear_model``` from ```sklearn```   </span>
* <span class="text-muted"> Create an instance ```LinearRegression()``` (called ```lr```)   </span>
* <span class="text-muted"> Fit your instance on ```X``` and ```y``` </span>
* <span class="text-muted"> Predict your data in ```ypred``` </span>

````python
from sklearn import linear_model
lr = linear_model.LinearRegression()
# training the model
lr.fit(X, y)
# making predictions
ypred = lr.predict(X)
````

* We can see coefficients:

````python
print(lr.coef_)
````

* We can calcule R2, MSE and RMSE:

````python
from sklearn.metrics import r2_score, mean_squared_error
R2 = r2_score(y, ypred)
MSE = mean_squared_error(y, ypred)
RMSE = np.sqrt(MSE)
print('R2 = ', R2, '\nMSE = ', MSE, '\nRMSE = ', RMSE)
````
